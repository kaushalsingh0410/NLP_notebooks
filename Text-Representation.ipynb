# Bag of Words
Bag of Words (BoW) model is a simple and widely used technique for text representation. It converts text into a numerical format by treating a document as a collection (bag) of its words, which keep total occurence of each words disregarding grammar, word order, and semantics.

### Probles with Word of Beg

#### 1. Ignores Word Order (Context)
The BoW model treats a text as a collection of individual words, disregarding the order in which they appear. This means it cannot distinguish between sentences like:

"The dog bit the man."

"The man bit the dog."

#### 2. No Semantic Understanding
Example: "buy" and "purchase" are treated as separate features, even though they mean the same.

#### 3. High Dimensionality
The dimensionality of the feature space grows with the size of the vocabulary. For a large corpus, this results in:

A sparse matrix (most elements are zeros).

Computational inefficiency in storage and processing.

#### 4. Cannot Handle Out-of-Vocabulary Words
BoW does not handle new or unseen words effectively because the vocabulary is fixed during training.

#### 5. Lack of Weighting Mechanism
BoW assigns equal importance to all words, including stop words like "the", "is", or "and".

# Word Embedding
Word embedding is a technique in Natural Language Processing (NLP) where words are represented as dense vectors in a continuous vector space. These embeddings capture semantic and syntactic relationships between words based on their context and usage in a large corpus of text.



# How Word Embedding Works
Word embeddings are typically learned using neural networks or probabilistic models on large corpora. The main idea is to learn a low-dimensional representation (e.g., 50-300 dimensions) such that similar words are close to each other in this vector space.

## Word2Vec
Word2Vec is a popular word embedding technique developed by Google. It represents words as dense vectors in a continuous vector space, capturing semantic and syntactic relationships between words. Word2Vec has two main architectures:

Continuous Bag of Words (CBOW): Predicts the current word based on the surrounding context words.

Skip-Gram: Predicts the surrounding context words given the current word.

## CBOW
CBOW is a neural network-based algorithm that predicts a target word given its surrounding context words

Example: For the sentence "The cat sat on the mat", CBOW uses the context ["The", "sat", "on"] to predict the target word cat.

Skip-Gram works exactly opposite to CBOW. It takes the center word as input and the remaining context words as output.
